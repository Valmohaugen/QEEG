```python
# Approach 2: Using sum() on boolean masks
depress_count = (df['specific.disorder'] == 'Depressive disorder').sum()
normal_count = (df['specific.disorder'] == 'Healthy control').sum()

print(f"Depress: {depress_count}")
print(f"Normal: {normal_count}")
```

```python
top_features = [
    'COH.delta.F7.T3',
    'COH.theta.F7.T3',
    'COH.theta.F7.C3',
    'COH.theta.F7.T5',
    'COH.theta.F7.O1',
    'COH.alpha.F7.T3',
    'COH.alpha.F7.T5',
    'COH.alpha.F7.O1',
    'COH.beta.F7.T5',
    'COH.beta.F7.O1'
]
df_top = df[top_features]
df['specific.disorder'] = df['specific.disorder'].map({
    'Depressive disorder': 1,
    'Healthy control': 0
})
target = df['specific.disorder'].to_list()
assert len(df_top), len(target)
print("Total Dataset length", len(df_top))
```

```python
#Normalization of Data
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
data = scaler.fit_transform(df_top)
data[0]
```

```python
#train-test split
from sklearn.model_selection import train_test_split

# Assuming X is your features DataFrame and y is your label Series
X_train, X_test, y_train, y_test = train_test_split(
    data, target, test_size=0.2, random_state=42, stratify=target
)
```

```python
import pandas as pd

# Combine training features and labels into a DataFrame
train_df = pd.DataFrame(X_train)
train_df['target'] = y_train  # Add the target column

from pycaret.classification import *
clf_setup = setup(data=train_df, target='target', session_id=42)
best_model = compare_models()

```

```python
from sklearn.metrics import accuracy_score, classification_report

# Combine test features and labels into DataFrame
test_df = pd.DataFrame(X_test)
test_df['target'] = y_test

# Predict and evaluate
preds = predict_model(best_model, data=test_df)

# Calculate accuracy
accuracy = accuracy_score(preds['target'], preds['prediction_label'])
print(f"Accuracy on test set: {accuracy}")

# Get all classification scores
report = classification_report(preds['target'], preds['prediction_label'])
print("Classification Report:\n", report)
```

```python
import random
import json

def generate_random_supplementary_info(prediction_score):
    # Simulate predicted label
    predicted_label = "Depressive disorder" if prediction_score > 0.8 else "normal"

    supplementary_info = {
        "mood": {
            "sadness": random.choice([True, False]),
            "irritability": random.choice([True, False]),
            "mood_swings": random.choice([True, False])
        },
        "sleep": {
            "insomnia": random.choice([True, False]),
            "hypersomnia": random.choice([True, False])
        },
        "energy_levels": random.choice(["low", "normal", "high"]),
        "cognitive_symptoms": {
            "concentration_difficulty": random.choice([True, False]),
            "memory_issues": random.choice([True, False])
        },
        "behavioral_changes": {
            "social_withdrawal": random.choice([True, False]),
            "reduced_activity": random.choice([True, False])
        },
        "physical_symptoms": {
            "appetite_change": random.choice([True, False]),
            "body_pain": random.choice([True, False])
        },
        "risk_factors": {
            "family_history_depression": random.choice([True, False]),
            "recent_trauma": random.choice([True, False]),
            "substance_abuse": random.choice([True, False])
        }
    }

    # Final JSON structure
    output_json = {
        "prediction_score": round(prediction_score, 4),
        "predicted_label": predicted_label,
        "supplementary_info": supplementary_info
    }

    return output_json

# generated_json = generate_random_supplementary_info(pred_scores[0])

# Pretty print
# print(json.dumps(generated_json, indent=4))
```

```python
json_arrays= []
for pred in pred_scores:
  generated_json = generate_random_supplementary_info(pred)
  json_arrays.append(generated_json)

dump_data= {
    'data': json_arrays
}

# Save to file
with open('depression_predictions.json', 'w') as f:
    json.dump(dump_data, f, indent=4)
```
